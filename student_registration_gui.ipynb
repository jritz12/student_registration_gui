```python
import sqlite3

# connect to our database
conn = sqlite3.connect('C:/Users/jackr/sis_db_2700')

# drop the "with closing" since there is no need for a context manager with the cursor
c = conn.cursor()
user_sid = input("Enter a Student ID: ")
sql = '''SELECT studentCourse.sid, studentCourse.cid, Course.creditHours
        FROM Course
        JOIN studentCourse ON Course.cid = studentCourse.cid
        WHERE studentCourse.sid = ?'''
c.execute(sql, (user_sid,))

# get all rows from the SELECT query
rows = c.fetchall()

# display the selected rows
for row in rows:
    print(row)

# close the cursor and the database connection
c.close()
conn.close()
```

```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


# GUI coding for Student Registration function

# Import Module
from tkinter import *

# Function for clearing the contents of all entry boxes 
def clear_all() :

	# whole content of entry boxes is deleted 
	sid_field.delete(0, END) 
	cid_field.delete(0, END)

#  PLACE FUNCTION HERE TO PERFORM SQL INSERT

def student_search():
    global sid_field
    global sname_field
    value1 = sid_field.get()
   # value2 = sname_field.get()
    course1.config(text="")
    course2.config(text="")
    course3.config(text="")
    ch1.config(text="")
    ch2.config(text="")
    ch3.config(text="")
    
    
    conn = sqlite3.connect('C:/Users/jackr/sis_db_2700')
    cursor = conn.cursor()
    
    query = '''SELECT studentCourse.sid, studentCourse.cid, Course.creditHours
        FROM Course
        JOIN studentCourse ON Course.cid = studentCourse.cid
        WHERE studentCourse.sid = ?'''
    cursor.execute(query, (value1,))
    
    results = cursor.fetchall()
    
    conn.close()
        
    courses = [course1, course2, course3]
    credit_hours = [ch1, ch2, ch3]
    for i, result in enumerate(results[:3]):
        courses[i].config(text=result[1])
        credit_hours[i].config(text=result[2])
    

# create root window
root = Tk()

# root window title and dimension
root.title("Student Registration")
# Set geometry(widthxheight)
root.geometry('475x200')
root.resizable(0, 0)
# adding labels to the root window
lbl1 = Label(root, text = "Student ID")
lbl1.grid()
lbl2 = Label(root, text = "Student Name:")
lbl2.grid()
lbl3 = Label(root, text = "Course Title")
lbl3.grid()
lbl4 = Label(root, text = "Credit Hours")
lbl4.grid()

search_button = Button(root, text="Search Database", command=student_search)
search_button.grid(row=2, column=2, columnspan=2)

sid_field = Entry(root) 
sname_field = Entry(root)
course1 = Label(root,bg="white", width=20)
course2 = Label(root,bg="white", width=20)
course3 = Label(root,bg="white", width=20)

ch1= Label(root,bg="white", width=20)
ch2= Label(root,bg="white", width=20)
ch3= Label(root,bg="white", width=20)

sid_field.grid(row = 1, column = 1, padx = 10, pady = 10) 
sname_field.grid(row = 2, column = 1, padx = 10, pady = 10)


lbl1.grid(row = 1, column = 0, padx = 10, pady = 10) 
lbl2.grid(row = 2, column = 0, padx = 10, pady = 10)
lbl3.grid(row = 3, column = 0, padx = 10, pady = 10)

lbl4.grid(row = 3, column = 3, padx = 10, pady =10)

course1.grid(row = 4, column = 0, padx = 10, pady = 1)
course2.grid(row = 5, column = 0, padx = 10, pady = 1)
course3.grid(row = 6, column = 0, padx = 10, pady = 1)


ch1.grid(row = 4, column = 3, padx = 10, pady = 1)
ch2.grid(row = 5, column = 3, padx = 10, pady = 1)
ch3.grid(row = 6, column = 3, padx = 10, pady = 1)
# Execute Tkinter
root.mainloop()


# In[ ]:




```
